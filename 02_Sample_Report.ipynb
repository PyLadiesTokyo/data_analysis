# サンプル分析のレポート

## 実験概要

ほげほげの性能を改善するために分析を行った．  
サンプル x を○○して性能が△△ポイント改善することを確認した．  
サンプル x と y の関係式は以下の通り：

$$y=2x$$

```python
x = range(1, 10)
y = range(2, 20, 2)

# サンプル実験のグラフを描画

plt.plot(x, y, 'o--')
plt.title('サンプル実験の結果')
plt.xlabel('サンプル x の値')
plt.ylabel('サンプル y の値')
plt.show()
```
